```
#Ejemplo 3.5

num = float(input("Introduce el número: "))
while(num != -1):
  cub = num ** 3
  print(f"el cubo del numero {num} es: {cub}")
  num = float(input("Introduce el número: "))
print("Fin del programa UwU")
```

```
#Problema 3.1
sumpar=0
sumimp=0
cuepar=0
I=1
while (I <= 5):
  num=int(input("Dame un número"))
  if num != 0:
    if ((-1) ** num) > 0: #num %2 (módulo)
      print("Es par")  
      sumpar = sumpar + num
      cuepar = cuepar + 1
    else:
      sumimp = sumimp + num
  I = I + 1
propar = sumpar / cuepar
print(f"El promedio de los pares es: {propar}")
print(f"El promedio de los pares es: {sumimp}")
```
